Example of a simple iPython notebook to read and analyze GHOST data

```python
from numpy import *
from glob import glob
from matplotlib.pyplot import *
%matplotlib inline
```

```python
path = '../../3D/bin/'
```

```python
t, ene, ens, eps = np.loadtxt(path+'balance.txt',unpack=True)
plot(t,ene)
xlabel('time')
ylabel('Energy')
```

```python
filelist = glob(path+'kspectrum.*.txt')
nfiles = size(filelist)
for i in range(1, nfiles, 5):
  ene = loadtxt(filelist[i])
  loglog(ene[:,0],ene[:,1])
nmax = 2*(size(ene[:,0])-1)
plt.xlim(1,nmax/3)
plt.xlabel('k')
plt.ylabel('E(k)')
```
